```python
import os
import numpy as np
import matplotlib as mpl
from matplotlib import pyplot as plt
from scipy.stats import poisson
from scipy.interpolate import InterpolatedUnivariateSpline

# colors
prop_cycle = plt.rcParams['axes.prop_cycle']
colors = prop_cycle.by_key()['color']

# average NPE for signal (Cherenkov) and background (ambient photons)
sig_mu = 10
bkg_mu = 2

# number of samples for fast-sim
Nsamples = 1000000
```

```python
# creating a numpy array for x-axis
# using step size as 1
x = np.linspace(0, 20, 21)
# finer bins for smooth curves
xsp = np.linspace(0, 20, 101)
# add several more bins at the end for histogram to get rid of extrapolation unstability (technical issue)
xbins = np.linspace(0, 20 + 5, 21 + 5)

# get the distributions (and smooth them with the spline fit)
# poisson distribution for background
y_bkg = poisson.pmf(x, mu=bkg_mu)
ysp_bkg = InterpolatedUnivariateSpline(x, y_bkg)(xsp)

# poisson distribution for signal (almost Gaussian)
y_sig = poisson.pmf(x, mu=sig_mu)
ysp_sig = InterpolatedUnivariateSpline(x, y_sig)(xsp)

# uncorrelated sum
ysp_sum = ysp_bkg + ysp_sig
```

```python
# plot the uncorrelated sum of background and signal
fig, ax = plt.subplots(figsize=(8, 6))
ax.plot(xsp, ysp_bkg, label='True Background', color=colors[1], ls='--')
ax.fill_between(xsp, 0, ysp_bkg, color=colors[1], alpha=0.2)
ax.plot(xsp, ysp_sig, label='True Signal', color=colors[2], ls='--')
ax.fill_between(xsp, ysp_bkg, ysp_sum, color=colors[2], alpha=0.2)
ax.plot(xsp, ysp_sum, label='Measurement (Uncorr. Sum)', color='k', ls='-', lw=2)
ax.set_axisbelow(True)
ax.tick_params(labelsize=18)
ax.set_ylabel('Normalized Distribution', fontsize=18)
ax.set_xlabel('NPE', fontsize=18)
ax.legend(fontsize=18)
ax.set_xlim(0)
ax.set_ylim(0)

# showing the graph
plt.show()
```

```python
# correlated sum
# assuming every event is signal + background (1:1 ratio, 100% pile-up)
samp_bkg = poisson.rvs(bkg_mu, size=Nsamples)
samp_sig = poisson.rvs(sig_mu, size=Nsamples)
# correlated sum
# note: corr. sum happens at event level (signal and background are merged event-wise, or piled-up),
#       while uncorr. sum happens at an aggregated level (spectrum addition)
samp_sum = samp_bkg + samp_sig

# get the distributions (and smooth them with the spline fit)
hvals_sum, _ = np.histogram(samp_sum, bins=xbins)
hvsp_sum = InterpolatedUnivariateSpline(xbins[:-1], hvals_sum/np.sum(hvals_sum))(xsp)

hvals_sig, _ = np.histogram(samp_sig, bins=xbins)
hvsp_sig = InterpolatedUnivariateSpline(xbins[:-1], hvals_sig/np.sum(hvals_sig))(xsp)

hvals_bkg, _ = np.histogram(samp_bkg, bins=xbins)
hvsp_bkg = InterpolatedUnivariateSpline(xbins[:-1], hvals_bkg/np.sum(hvals_bkg))(xsp)

# plot
fig, ax = plt.subplots(figsize=(8, 6))
ax.plot(xsp, hvsp_bkg, label='True Background', color=colors[1], ls='--')
ax.fill_between(xsp, 0, hvsp_bkg, color=colors[1], alpha=0.2)
ax.plot(xsp, hvsp_sig, label='True Signal', color=colors[2], ls='--')
ax.fill_between(xsp, 0, hvsp_sig, color=colors[2], alpha=0.2)
ax.plot(xsp, hvsp_sum, label='Measurement (Corr. Sum)', color='k', ls='-', lw=2)
ax.set_axisbelow(True)
ax.tick_params(labelsize=18)
ax.set_ylabel('Normalized Distribution', fontsize=18)
ax.set_xlabel('NPE', fontsize=18)
ax.legend(fontsize=18)
ax.set_xlim(0)
ax.set_ylim(0)

# showing the graph
plt.show()
```

```python
# showing that a fully corr. sum - the mean value of background -> the mean value of signals
fig, ax = plt.subplots(figsize=(8, 6))
ax.plot(xsp, hvsp_sig, label='True Signal', color=colors[2], ls='--')
ax.fill_between(xsp, 0, hvsp_sig, color=colors[2], alpha=0.2)
ax.plot(xsp - bkg_mu, hvsp_sum, label=r'Measurement - $\mu_{BKG}$', color='k', ls='-', lw=2)
ax.set_axisbelow(True)
ax.tick_params(labelsize=18)
ax.set_ylabel('Normalized Distribution', fontsize=18)
ax.set_xlabel('NPE', fontsize=18)
ax.legend(loc='lower center', fontsize=18)
ax.set_xlim(0)
ax.set_ylim(0)
ax.text(0.95, 0.95, '100% piled-up', fontsize=22, transform=ax.transAxes, va='top', ha='right')

# showing the graph
plt.show()
```

```python
# what if pile-up probability is not 100%?
# For example: 20%, 40%, 60%, 80%
# NOTE that the distributions of signals and backgrounds are not changed
pup_probs = [0.2, 0.4, 0.6, 0.8]
# a uniform [0., 1.] distribution for the pile-up probability sampling
pup = np.random.uniform(0., 1., Nsamples)

fig, axs = plt.subplots(2, 2, figsize=(12, 9), sharex='all', sharey='all', gridspec_kw=dict(hspace=0., wspace=0.))

for i, (pup_prob, ax) in enumerate(zip(pup_probs, axs.flat)):
    # part of the signal is piled-up with background
    samp_sum = samp_sig + (pup <= pup_prob)*samp_bkg
    # get the distributions (and smooth them with the spline fit)
    hvals_sum, _ = np.histogram(samp_sum, bins=xbins)
    hvsp_sum = InterpolatedUnivariateSpline(xbins[:-1], hvals_sum/np.sum(hvals_sum))(xsp)
    
    hvals_sig, _ = np.histogram(samp_sig, bins=xbins)
    hvsp_sig = InterpolatedUnivariateSpline(xbins[:-1], hvals_sig/np.sum(hvals_sig))(xsp)
    
    hvals_bkg, _ = np.histogram(samp_bkg, bins=xbins)
    hvsp_bkg = InterpolatedUnivariateSpline(xbins[:-1], hvals_bkg/np.sum(hvals_bkg))(xsp)
    
    # plot the subtracted result
    ax.plot(xsp, hvsp_sig, label='True Signal', color=colors[2], ls='--')
    ax.fill_between(xsp, 0, hvsp_sig, color=colors[2], alpha=0.2)
    ax.plot(xsp - bkg_mu, hvsp_sum, label=r'Measurement - $\mu_{BKG}$', color='k', ls='-', lw=2)
    ax.set_axisbelow(True)
    ax.tick_params(labelsize=18)
    ax.set_xlim(0)
    ax.set_ylim(0)
    ax.text(0.95, 0.95, '{}%\npiled-up'.format(pup_prob*100.), fontsize=22, transform=ax.transAxes, va='top', ha='right')
    if i == 0:
        ax.legend(loc='lower center', fontsize=18)

fig.text(0.02, 0.5, 'Normalized Distribution', fontsize=18, va='center', rotation=90)
fig.text(0.5, 0.02, 'NPE', fontsize=18, ha='center')
# showing the graph
plt.show()
```

```python
# now check the single-photon (mu = 1) pile-up effects, for both true backgrounds and signals
samp_spe = poisson.rvs(1, size=Nsamples)

samp_sig_pup = samp_sig + samp_spe
samp_bkg_pup = samp_bkg + samp_spe

hvals_sig_pup, _ = np.histogram(samp_sig_pup, bins=xbins)
hvsp_sig_pup = InterpolatedUnivariateSpline(xbins[:-1], hvals_sig_pup/np.sum(hvals_sig_pup))(xsp)

hvals_bkg_pup, _ = np.histogram(samp_bkg_pup, bins=xbins)
hvsp_bkg_pup = InterpolatedUnivariateSpline(xbins[:-1], hvals_bkg_pup/np.sum(hvals_bkg_pup))(xsp)

# 100% pile-up
fig, axs = plt.subplots(1, 2, figsize=(12, 5), sharex='all', sharey='all', gridspec_kw=dict(hspace=0., wspace=0.))

# plot data
ax = axs[0]
ax.plot(xsp, hvsp_sig, label='Signal', color=colors[2], ls='--')
ax.fill_between(xsp, 0, hvsp_sig, color=colors[2], alpha=0.2)
ax.plot(xsp, hvsp_sig_pup, label=r'Signal + SPE', color='k', ls='-', lw=2)

# plot data
ax = axs[1]
ax.plot(xsp, hvsp_bkg, label='Background', color=colors[1], ls='--')
ax.fill_between(xsp, 0, hvsp_bkg, color=colors[1], alpha=0.2)
ax.plot(xsp, hvsp_bkg_pup, label=r'Background + SPE', color='k', ls='-', lw=2)

# axis setup
for i, ax in enumerate(axs.flat):
    ax.set_axisbelow(True)
    ax.tick_params(labelsize=16)
    ax.set_xlim(0)
    ax.set_ylim(0)
    # ax.text(0.95, 0.95, '{}%\npiled-up'.format(100), fontsize=22, transform=ax.transAxes, va='top', ha='right')
    ax.legend(title='100% piled-up', loc='upper right', fontsize=18, title_fontsize=18)


fig.text(0.03, 0.5, 'Normalized Distribution', fontsize=18, va='center', rotation=90)
fig.text(0.5, 0.01, 'NPE', fontsize=18, ha='center')
# showing the graph
plt.show()
```

```python
# scan over pile-up probability
pup_probs = np.linspace(0, 1.0, 11)
pup_sig = np.random.uniform(0., 1., Nsamples)
pup_bkg = np.random.uniform(0., 1., Nsamples)

data = []
for pup_prob in pup_probs:
    samp_sig_pup = samp_sig + (pup_sig <= pup_prob)*samp_spe
    samp_bkg_pup = samp_bkg + (pup_bkg <= pup_prob)*samp_spe
    data.append((pup_prob, np.mean(samp_sig_pup) - np.mean(samp_sig), np.mean(samp_bkg_pup) - np.mean(samp_bkg)))
data = np.array(data)

# mean-value difference plot
fig, ax = plt.subplots(figsize=(8, 6))
ax.plot(data.T[0]*100., data.T[1], 'o', label=r'$\mu_{SIG}^{HR}$ - $\mu_{SIG}$', ms=10, fillstyle='none')
ax.plot(data.T[0]*100., data.T[2], 'v', label=r'$\mu_{BKG}^{HR}$ - $\mu_{BKG}$', ms=10)
ax.set_axisbelow(True)
ax.tick_params(labelsize=16)
ax.set_xlim(0)
ax.set_ylim(0)

ax.legend(loc='upper left', fontsize=18)
ax.set_xlabel('Pile-up Probability (%)', fontsize=18)
plt.show()
```
